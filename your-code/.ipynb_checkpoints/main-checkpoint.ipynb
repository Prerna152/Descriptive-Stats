## Challenge 1
#### 1.- Define a function that simulates rolling a dice 10 times. Save the information in a dataframe.
**Hint**: you can use the *choices* function from module *random* to help you with the simulation.

```python
# your code here
def roll_dice_10_times():
    # Define the possible outcomes for a dice roll
    dice_faces = [1, 2, 3, 4, 5, 6]
    
    # Simulate rolling the dice 10 times
    rolls = choices(dice_faces, k=10)
    
    # Create a DataFrame to store the roll outcomes
    # Use roll number as index starting from 1 for a more intuitive understanding
    rolls_df = pd.DataFrame({'Roll': range(1, 11), 'Outcome': rolls})
    
    return rolls_df

# Call the function and display the results
dice_rolls_df = roll_dice_10_times()
print(dice_rolls_df)
```

 2.- Plot the results sorted by value.

```python
# your code here
dice_rolls_df = roll_dice_10_times().sort_values(by='Outcome')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(dice_rolls_df['Roll'].astype(str), dice_rolls_df['Outcome'], color='skyblue')
plt.title('Dice Roll Outcomes Sorted by Value')
plt.xlabel('Roll Number')
plt.ylabel('Outcome')
plt.xticks(rotation=45)  
plt.grid(axis='y', alpha=0.75)  

plt.show()
```

#### 3.- Calculate the frequency distribution and plot it. What is the relation between this plot and the plot above? Describe it with words.

```python
# your code here
frequency_distribution = dice_rolls_df['Outcome'].value_counts().sort_index()
```

```python
"""
your comments here
"""
# Plotting the frequency distribution
frequency_distribution.plot(kind='bar', color='skyblue')
plt.title('Frequency Distribution of Dice Rolls')
plt.xlabel('Dice Outcome')
plt.ylabel('Frequency')
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal for readability
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Challenge 2
Now, using the dice results obtained in *challenge 1*, your are going to define some functions that will help you calculate the mean of your data in two different ways, the median and the four quartiles. 

#### 1.- Define a function that computes the mean by summing all the observations and dividing by the total number of observations. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here
def calculate_mean(observations):
    # Calculate the sum of all observations
    total_sum = sum(observations)
    
    # Calculate the number of observations
    num_observations = len(observations)
    
    # Calculate the mean by dividing the total sum by the number of observations
    mean = total_sum / num_observations
    
    return mean

# Assuming dice_rolls is a list of  dice roll outcomes
# For demonstration, let's use a sample list of dice roll outcomes
dice_rolls = [1, 3, 4, 6, 3, 4, 2, 5, 6, 2]

# Call the function and print the result
mean_dice_rolls = calculate_mean(dice_rolls)
print(f"The mean of the dice rolls is: {mean_dice_rolls}")
```

#### 2.- First, calculate the frequency distribution. Then, calculate the mean using the values of the frequency distribution you've just computed. You are not allowed to use any methods or functions that directly calculate the mean value. 

```python
# your code here# 
dice_rolls = [1, 3, 4, 6, 3, 4, 2, 5, 6, 2]  

#  creating frequency distribution
frequency_distribution = {}
for outcome in dice_rolls:
    if outcome in frequency_distribution:
        frequency_distribution[outcome] += 1
    else:
        frequency_distribution[outcome] = 1
```

```python
# Sum product of values and frequencies
sum_product = sum(value * frequency for value, frequency in frequency_distribution.items())

# Sum of all frequencies (total number of observations)
total_observations = sum(frequency_distribution.values())

# Calculate mean
mean_value = sum_product / total_observations

print(f"The mean from the frequency distribution is: {mean_value}")
```

 3.- Define a function to calculate the median. You are not allowed to use any methods or functions that directly calculate the median value. 
**Hint**: you might need to define two computation cases depending on the number of observations used to calculate the median.

```python
# your code here
def calculate_median(data):
    # Sort the data
    sorted_data = sorted(data)
    
    # Calculate the number of observations
    N = len(sorted_data)
    
    # Determine if the number of observations is odd or even
    is_odd = N % 2 != 0
    
    if is_odd:
        # If odd, the median is the value at the middle position
        median_position = N // 2
        median_value = sorted_data[median_position]
    else:
        # If even, the median is the average of the two middle values
        median_position_1 = (N // 2) - 1  # Position is subtracted by 1 because list index starts at 0
        median_position_2 = N // 2
        median_value = (sorted_data[median_position_1] + sorted_data[median_position_2]) / 2
    
    return median_value

# Example usage with odd and even number of data points
data_odd = [1, 3, 3, 6, 7, 8, 9]
data_even = [1, 2, 3, 4, 5, 6, 8, 9]

print(f"Median (Odd):  {calculate_median(data_odd)}")
print(f"Median (Even): {calculate_median(data_even)}")
```

#### 4.- Define a function to calculate the four quartiles. You can use the function you defined above to compute the median but you are not allowed to use any methods or functions that directly calculate the quartiles. 

```python
# your code here
def calculate_median(data):
    sorted_data = sorted(data)
    N = len(sorted_data)
    middle_point = N // 2
    
    if N % 2 == 0:  # Even number of observations
        median_value = (sorted_data[middle_point - 1] + sorted_data[middle_point]) / 2
    else:  # Odd number of observations, exclude the middle value for the halves
        median_value = sorted_data[middle_point]
    
    return median_value

def calculate_quartiles(data):
    sorted_data = sorted(data)
    N = len(sorted_data)
    
    Q2 = calculate_median(sorted_data)  
    
    if N % 2 == 0:
        lower_half = sorted_data[:N // 2]
        upper_half = sorted_data[N // 2:]
    else:
        lower_half = sorted_data[:N // 2]  # Up to the median, exclusive
        upper_half = sorted_data[N // 2 + 1:]  # After the median, exclusive
    
    Q1 = calculate_median(lower_half)
    Q3 = calculate_median(upper_half)
    Q4 = sorted_data[-1]  # Maximum value, or 100% of the data spread
    
    return Q1, Q2, Q3, Q4

# Example usage
data = [7, 15, 36, 39, 40, 41]
Q1, Q2, Q3, Q4 = calculate_quartiles(data)
print(f"Q1: {Q1}, Q2 (Median): {Q2}, Q3: {Q3}, Q4 (Max): {Q4}")
```
